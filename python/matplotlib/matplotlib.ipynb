### 前面的代码是先创建画布然后创建图形矩阵

```python
# 创建图形矩阵的快捷函数
# sharex,sharey共享x,y轴，也就是刻度线一样
fig,axes = plt.subplots(2,2,sharex=True,sharey=True)
for i in range(2):
    for j in range(2):
        axes[i,j].hist(randn(500),bins=50,color='k',alpha=0.5)
plt.subplots_adjust(wspace=0,hspace=0)
```

```python
# 积分图
# 其实是绘制多边形图实现积分图
import matplotlib.pyplot as plt
from matplotlib.patches import Polygon#多边形
import numpy as np
def func(x):
    return -(x-2)*(x-8)+40
x=np.linspace(0,10)
y=func(x)

fig.ax=plt.subplot()
plt.plot(x,y,'r',linewidth=2)
a,b=2,9
ax.set_xticks([a,b])
# ax.set_yticks([])
ax.set_xticklabels(['$a$','$b$'])
```

```python
# 绘制积分区域的多边形点位置
ix=np.linspace(a,b)
iy=func(ix)
ixy=zip(ix,iy)
verts=[(a,0)]+list(ixy)+[(b,0)]
poly=Polygon(verts,facecolor='0.9',edgecolor='0.1')
ax.add_patch(poly)

plt.figtext(0.9,0.05,'$x$')
plt.figtext(0.1,0.9,'$y$')

x_match=(a+b)*0.4
y_match=30
plt.text(x_match,y_match,
        r'$inta^b (-(x-2)*(x-8)+40)dx$',fontsize=20)

plt.show()
```
