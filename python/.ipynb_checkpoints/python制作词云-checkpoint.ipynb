### python 制作词云图

```python
# -*- coding=utf-8 -*-
'''
@author:Roc-J
@time:2017-08-30
'''
from os import path
from wordcloud import WordCloud
import matplotlib.pyplot as plt

d = path.dirname('__file__')

# 读取一个txt文件
text = open(path.join(d, 'constitution.txt')).read()

# 生成一个词云图像
wordcloud = WordCloud().generate(text)
# 显示展示的图像
plt.imshow(wordcloud)
plt.axis("off")

# lower max_font_size
wordcloud = WordCloud(min_font_size=30,max_font_size=60).generate(text)
plt.figure()
plt.imshow(wordcloud)
plt.axis("off")
plt.show()
```

词云展示  
上面展示的是英文，如果显示中文还需要其他设置

```python
# -*- coding: utf-8 -*-
#wordcloud生成中文词云

from wordcloud import WordCloud
import codecs
import jieba
from scipy.misc import imread
import os
from os import path
import matplotlib.pyplot as plt
from PIL import Image, ImageDraw, ImageFont
```

```python
# 绘制词云
def draw_wordcloud():
    #读入一个txt文件
    comment_text = open('./lyrics.txt','r').read()
    #分词，生成字符串，如果不通过分词，无法直接生成正确的中文词云
    cut_text = " ".join(jieba.cut(comment_text))
    # d = path.dirname('__file__') # 当前文件文件夹所在目录
    color_mask = imread("test.jpg") # 读取背景图片
    cloud = WordCloud(
        #设置字体，不指定就会出现乱码
        font_path="./HYQiHei-25JF.ttf",
        #font_path=path.join(d,'simsun.ttc'),
        #设置背景色
        background_color='white',
        #词云形状
        mask=color_mask,
        #允许最大词汇
        max_words=2000,
        #最大号字体
        max_font_size=80
    )
    word_cloud = cloud.generate(cut_text) # 产生词云
    word_cloud.to_file("pjl_cloud4.jpg") #保存图片
    #  显示词云图片
    plt.imshow(word_cloud)
    plt.axis('off')
    plt.show()
```

```python
draw_wordcloud()
```
