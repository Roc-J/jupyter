```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
np.random.seed(sum(map(ord,'aesthetics')))
```

```python
# 定义一个函数用来画正弦函数，这将帮助我们了解我们可以控制的不同风格的参数
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x,np.sin(x+i*.5)*(7-i)*flip)
```

```python
sinplot()
```

### 转换成Seaborn模块，只需要引入seaborn模块

```python
import seaborn as sns
# to switch to seaborn defaults,simply call the  set() function
sns.set()
sinplot()
```

请注意，在0.8之前的seaborn版本中，set()被调用导入，在更高版本中，必须明确调用

Seaborn将matplotlib参数分为两个独立的组。第一组设定了绘图的的审美风格，第二组绘图的各种元素缩放，以便将其轻松并入不同的环境中。

用于操作这些参数的界面是两对函数。要控制风格，请使用axes_style()和set_style()功能。要缩放图形大小，请使用plotting_context()和set_context()功能。在这两种情况下，第一个函数返回参数字典，第二个函数设置matplotlib默认值。

### Seaborn figure style  
* darkgrid 
* whitegrid
* dark
* white 
* ticks

the default theme is darkgrid  

```python
# plot the bcxplot
sns.set_style('whitegrid')
data = np.random.normal(size=(20,6)) + np.arange(6)/2
sns.boxplot(data=data)
```

```python
# for many plots the grid is less necessary
sns.set_style('dark')
sinplot()
```

```python
sns.set_style('white')
sinplot()
```

```python
# sometimes want to give a little ectra structure to the plots,so select the ticks
sns.set_style('ticks')
sinplot()
```

```python
# both white and dark style having the toping and right axes
# removing these by despine() function
sinplot()
sns.despine()
```

```python
# use trim limit the range
f,ax = plt.subplots()
sns.violinplot(data=data)
sns.despine(offset=10,trim=True)
```

```python
# control which spines removed with additional arguments to despine()
sns.set_style('whitegrid')
sns.boxplot(data=data,palette='deep')
sns.despine(left=True)
```

```python
# Temporarily setting figure style 
# it's easy to switch back and forth,
# use the function axes_style in a with statement
with sns.axes_style('darkgrid'):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

```python
# get the style
sns.axes_style()
```

```python
# user set the style
sns.set_style("darkgrid", {"axes.facecolor": ".9"})
sinplot()
```
